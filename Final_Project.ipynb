```python
continent_avg_iq = df.groupby('Continent')['IQ'].mean().reset_index()
plt.figure(figsize=(10, 6))
sns.barplot(x='Continent', y='IQ', data=continent_avg_iq, palette='Set2')
plt.title('Average IQ by Continent', fontsize=16)
plt.xlabel('Continent', fontsize=14)
plt.ylabel('Average IQ', fontsize=14)
for index, value in enumerate(continent_avg_iq['IQ']):
    plt.text(index, value, f'{value:.2f}', ha='center', va='bottom', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
